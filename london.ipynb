```python
missing_values = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing_Count': missing_values,
    'Missing_Percentage': missing_percentage
})

missing_df = missing_df[missing_df['Missing_Count'] > 0]
display(missing_df)

print(f"\nTotal missing values: {df.isnull().sum().sum()}")
```

```python
x = df.groupby('Neighborhood').agg({
	'Price (£)': ['min', 'max', 'mean'],
	'Bedrooms': ['min', 'max', 'mean'],
	'Bathrooms': ['min', 'max', 'mean'],
	'Square Meters': ['min', 'max', 'mean'],
	'Building Age': ['min', 'max', 'mean'],
	'Property Type': ['count', 'nunique']
})
x
```

```python
df["Price per meter"] = (df["Price (£)"] / df["Square Meters"]).round(2)
df
```

```python
df["Total Rooms"] = df["Bedrooms"] + df["Bathrooms"]
df
```

```python
df["Building Age"].agg(["min", "max", "mean"])
```

```python
mean_price = df['Price (£)'].mean()
print("Average Price (£):", mean_price)
```

```python
oldest_age = df['Building Age'].max()
oldest_houses = df[df['Building Age'] == oldest_age]
oldest_price_mean = oldest_houses['Price (£)'].mean()

print(f"Oldest House Age: {oldest_age} years")
print(f"Mean Price of Oldest House(s): £{oldest_price_mean:.2f}")
```

```python
newest_age = df['Building Age'].min()
newest_houses = df[df['Building Age'] == newest_age]
newest_price_mean = newest_houses['Price (£)'].mean()

print(f"Newest House Age: {newest_age} years")
print(f"Mean Price of Newest House(s): £{newest_price_mean:.2f}")
```

```python
df['Garden_Numeric'] = df['Garden'].map({'Yes': 1, 'No': 0})
print(df[['Garden', 'Garden_Numeric']].head())
```

```python
df['Garage_Numeric'] = df['Garage'].map({'Yes': 1, 'No': 0})
print(df[['Garage', 'Garage_Numeric']].head())
```

### Visualizing

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Price (£)'], bins=30, kde=True, color='lightcoral')
plt.title("Distribution of House Prices", fontsize=16)
plt.xlabel("Price (£)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

```python
sns.lineplot(x="Square Meters", y="Price (£)", hue="Neighborhood", data=df)
plt.title("Price vs Square Meters by Neighborhood")
plt.show()
```

```python
sns.barplot(x="Neighborhood", y="Price (£)", data=df, estimator=np.mean)
plt.title('Average Price by Neighborhood')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x="Neighborhood", y="Building Age", data=df)
plt.title('Distribution of Building Age by Neighborhood')
plt.xticks(rotation=45)
plt.show()
```

```python
correlation_matrix = df[['Bedrooms', 'Bathrooms', 'Square Meters', 'Building Age', 'Price (£)']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title("Correlation Matrix", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Balcony'], df['Square Meters'], marker='o', linestyle='-', color='teal')
plt.title("Balcony vs Square Meters", fontsize=16)
plt.xlabel("Balcony", fontsize=12)
plt.ylabel("Square Meters", fontsize=12)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Interior Style', palette='pastel', order=df['Interior Style'].value_counts().index)
plt.title("Distribution of Interior Styles", fontsize=16)
plt.xlabel("Interior Style", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid", palette="pastel")
sns.scatterplot(data=df, x='Building Age', y='Price (£)', hue='Neighborhood', style='Property Type', alpha=0.8, s=80)
plt.title("Building Age vs House Price", fontsize=16, fontweight='bold')
plt.xlabel("Building Age", fontsize=12)
plt.ylabel("Price (£)", fontsize=12)
plt.legend(title='Neighborhood', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
