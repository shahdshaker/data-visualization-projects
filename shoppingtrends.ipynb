```python
missing_values = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing_Count': missing_values,
    'Missing_Percentage': missing_percentage
})

missing_df = missing_df[missing_df['Missing_Count'] > 0]
display(missing_df)

print(f"\nTotal missing values: {df.isnull().sum().sum()}")
```

### Visualizing

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Purchase Amount (USD)"], bins=30, kde=True)
plt.title("Distribution of Purchase Amounts")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df, x="Age", hue="Gender", bins=20, kde=True, multiple="stack")
plt.title("Age Distribution by Gender")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x="Category", y="Purchase Amount (USD)", data=df, estimator="mean", ci=None)
plt.xticks(rotation=45)
plt.title("Average Purchase Amount per Category")
plt.show()
```

```python
sns.boxplot(x="Payment Method", y="Purchase Amount (USD)", data=df)
plt.title("Purchase Amount by Payment Method")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
sns.jointplot(x="Age", y="Purchase Amount (USD)", data=df, kind="hex", height=6, cmap="viridis")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(x="Preferred Payment Method", y="Age", data=df)
plt.xticks(rotation=45)
plt.title("Age Distribution by Preferred Payment Method")
plt.show()
```
