```python
missing_values = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing_Count': missing_values,
    'Missing_Percentage': missing_percentage
})

missing_df = missing_df[missing_df['Missing_Count'] > 0]
display(missing_df)

print(f"\nTotal missing values: {df.isnull().sum().sum()}")
```

```python
df["Postal Code"] = df.groupby("City")["Postal Code"].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x)
)
```

### Visualizing

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Sales"], bins=50, kde=True, color="teal")
plt.title("Sales Distribution")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x="Region", data=df)
plt.title("Orders Count by Region")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.barplot(x="Region", y="Sales", data=df, estimator=sum, palette="viridis")
plt.title("Total Sales by Region")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.barplot(x="Category", y="Sales", data=df, estimator=sum, palette="pastel")
plt.title("Total Sales by Category")
plt.show()
```

```python
sales_over_time = df.groupby("Order Date")["Sales"].sum().reset_index()
plt.figure(figsize=(12,5))
sns.lineplot(x="Order Date", y="Sales", data=sales_over_time)
plt.title("Sales Trend Over Time")
plt.show()
```

```python
sns.jointplot(x="Sales", y="Row ID", data=df, kind="scatter", alpha=0.5)
plt.suptitle("Sales vs. Order Size", y=1.02)
plt.show()
```

```python
top_subcats = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_subcats.values, y=top_subcats.index, palette="Blues_r")
plt.title("Top 10 Sub-Categories by Sales")
plt.xlabel("Total Sales")
plt.show()
```
